```python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.neural_network import MLPClassifier
import pandas as pd
```

## Obtendo os Dados

```python
d = datasets.load_iris()
# transform pandas dataframe
iris = pd.DataFrame(d.data, columns=d.feature_names)
```

```python
iris.info()
```

```python
iris.head()
```

## Treinando o modelo com o MLP Classifier

```python
X_train, X_test, y_train, y_test = train_test_split(d.data, d.target, test_size=0.3)
```

```python
clf = MLPClassifier(hidden_layer_sizes=(10,), max_iter=1500)
```

```python
clf.fit(X_train, y_train)
```

## MÃ©tricas de desempenho

```python
score = clf.score(X_test, y_test)
score
```

```python
predictions = clf.predict(X_test)
predictions
```

```python
classification_report(y_test, predictions)
```
